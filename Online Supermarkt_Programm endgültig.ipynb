```python
# Anna Dreier (262327), Norina Eisenbacher, Isabell Hinger (262330)
# Online Supermarkt

import pandas as pd
import ipywidgets as widgets
import time 

class color:
   LILA = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLAU = '\033[94m'
   GRÜN = '\033[92m'
   GELB = '\033[93m'
   ROT = '\033[91m'
   FETT = '\033[1m'
   UNTERSTRICHEN = '\033[4m'
   END = '\033[0m'


df = pd.read_csv('Liste_Gemüse.CSV', delimiter=';')
df.obst = pd.read_csv('Liste_Obst.csv', delimiter=';')

score = 0


#Angabe persönliche Daten
print("Willkommen im virtuellen Supermarkt! Du kannst dir Lebensmittel aussuchen und in den Warenkorb legen. \nNach Abschluss des Einkaufs, werden die Lebensmittel innerhalb eines Tages geliefert.")
print("Um eine reibungslose Ausstellung gewährleisten zu können, gebe bitte hier deine Kontaktdaten an:\n")

your_name = input('Dein Name: ')
your_city = input('Deine Stadt: ')
your_country = input('Dein Land: ')



class user():
    user = []
    def __init__(self, your_name, your_city, your_country):
        self.your_name = your_name
        self.your_city = your_city
        self.your_country= your_country
    def __repr__(self):
        return "Name: " + self.your_name + "\nStadt: " + self.your_city + "\nLand: " + self.your_country
user_1 = user(your_name, your_city, your_country)




#Angabe Land
if (your_country == 'Deutschland' or your_country == 'deutschland' or your_country == 'germany' or your_country == 'Germany' or your_country == 'de'):
    print("Vielen Dank,",your_name,"!")
    time.sleep(2)
else:
    print("Oh tut uns Leid, leider können wir keine Lebensmittel ins Ausland liefern! Eine Abholung in Deutschland ist allerdings möglich.")
    time.sleep(2)
print("\nFüge einfach die Lebensmittel hinzu, die du gerne möchtest. Uns liegt Nachhaltigkeit sehr am Herzen.\nWir würden uns deshalb freuen, wenn du bei deinem Einkauf darauf achtest.")


#Angabe Monat
print("Um die Nachhaltigkeit der Lebensmittel besser bewerten zu können, benötigen wir den Monat, in welchem du einkaufst.")

month = input("Bitte gebe hier den Monat ein:")
while month not in ['Januar', 'Februar', 'März', 'April', 'Mai', 'Juni', 'Juli', 'August', 'September', 'Oktober', 'November', 'Dezember']:
    print ("Bitte gebe den Monat ausgeschrieben an. Achte auf Groß- und Kleinschreibung. Beispiel: Januar")
    month = input("\nBitte gebe hier den Monat ein:")


einkaufsliste = []


#Gemüseregal
time.sleep(2)
print(color.GRÜN + color.FETT + "\nHier siehst du unser Gemüsesortiment!\n" + color.END)
print(df[['Lebensmittel', 'Herkunft']])


gemuesenummer=999
while (gemuesenummer >14 or gemuesenummer <0):
    gemuesenummer = input("\nGebe die Zahl des gewünschten Lebensmittels ein: ")
    try:
        gemuesenummer = int(gemuesenummer)
    except:
        print ("Ungütlige Eingabe")
        gemuesenummer = 999

gemuesenummer = int(gemuesenummer)
gemuese = df.iloc[gemuesenummer]
print("\nDu hast erfolgreich 1 ", gemuese[0], "deinem Einkaufskorb hinzugefügt")
einkaufsliste.append(gemuese[0])
score += (df.loc[gemuesenummer][month])        
      


i = "+"
weitergemuese = input("Wenn du ein weiteres Gemüse einkaufen möchtest, tippe auf das + oder, wenn du fertig bist, drücke Enter:")
while weitergemuese == "+":
    gemuesenummer=999
    while (gemuesenummer >14 or gemuesenummer <0):
        gemuesenummer = input("\nGebe die Zahl des gewünschten Lebensmittels ein: ")
        try:
            gemuesenummer = int(gemuesenummer)
        except:
            print ("Ungültige Eingabe")
            gemuesenummer = 999
    gemuesenummer = int(gemuesenummer)
    gemuese = df.iloc[gemuesenummer]

    print("\nDu hast erfolgreich 1 ", gemuese[0], "dem Warenkorb hinzugefügt")
    einkaufsliste.append(gemuese[0])
    weitergemuese = input("Wenn du ein weiteres Gemüse einkaufen möchtest, tippe auf das + oder, wenn du fertig bist, drücke Enter:")

    score += (df.loc[gemuesenummer][month]) 
    
#Nachhaltigkeitsscore
print("\nIm Moment hast du von 12 möglichen Nachhaltigkeitspunkten", score/len(einkaufsliste),"erreicht.")
print("Deine Einkaufsliste beinhaltet bis jetzt folgende Lebensmittel:\t", einkaufsliste)


#Obstregal
time.sleep(2)
print(color.GRÜN + color.FETT + "\nHier siehst du unser Obstsortiment!\n" + color.END)
print("Welche Lebenmittel möchtest du hier einkaufen?\n") 
print(df.obst[['Lebensmittel', 'Herkunft']])


obstnummer=999
while (obstnummer >14 or obstnummer <0):
    obstnummer = input("\nGebe die Zahl des gewünschten Lebensmittels ein: ")
    try:
        obstnummer = int(obstnummer)
    except:
        print ("Ungültige Eingabe")
        obstnummer = 999

obstnummer = int(obstnummer)
obst = df.obst.iloc[obstnummer]
score += (df.obst.loc[obstnummer][month])


print("\nDu hast erfolgreich 1 ", obst[0], "dem Warenkorb hinzugefügt")
einkaufsliste.append(obst[0])




i = "+"
weiterobst = input("Wenn du ein weiteres Obst einkaufen möchtest, tippe auf das + oder, wenn du fertig bist, drücke Enter:")
while weiterobst == "+":
    
    obstnummer=999
    while (obstnummer >14 or obstnummer <0):
        obstnummer = input("\nGebe die Zahl des gewünschten Lebensmittels ein: ")
        try:
            obstnummer = int(obstnummer)
        except:
            print ("Ungültige Eingabe")
            obstnummer = 999

    obstnummer = int(obstnummer)
    obst = df.obst.iloc[obstnummer]

    print("\nDu hast erfolgreich 1 ", obst[0], "dem Warenkorb hinzugefügt")
    einkaufsliste.append(obst[0])
    weiterobst = input("Wenn du ein weiteres Obst einkaufen möchtest, tippe auf das + oder, wenn du fertig bist, drücke Enter:")

    score += (df.obst.loc[obstnummer][month])

# Überprüfung der persönlichen Daten
print ("\nFolgende Angaben werden für die Lieferung verwendet: ")
print(user_1)

#Überprüfung des Warenkorbs
time.sleep(1)
print ("\nHier siehst du nochmal deine ausgewählten Lebensmittel: ", einkaufsliste)
time.sleep(1)

#Bewertung des Warenkorbs nach Nachhaltigkeit
if score/len(einkaufsliste) > 8:
    print (color.GRÜN + "\nDu hast von 12 Nachhaltigkeitspunkten", score/len(einkaufsliste), "erreicht.\nSuper! Dein Einkauf war sehr nachhaltig. Weiter so." + color.END)
elif score/len(einkaufsliste) < 4:
    print (color.ROT + "\nDu hast von 12 Nachhaltigkeitspunkten", score/len(einkaufsliste), "erreicht.\nNächstes Mal kannst du versuchen besser auf deine Umwelt zu achten und die Nachhaltigkeit besser im Auge zu behalten." + color.END)
else:
    print (color.GELB + "\nDu hast von 12 Nachhaltigkeitspunkten", score/len(einkaufsliste), "erreicht.\nToll, dass du bereits auf die Nachhaltigkeit geachtet hast. Aber geht das noch besser?" + color.END)

#Bestätigung des Einkaufs
time.sleep(1)
print ("\nWenn du mit deinem Einkauf zufrieden bist, drücke bitte auf bestätigen.")

from IPython.display import display
button = widgets.Button(description="Bestellung bestätigen")
output = widgets.Output()

display(button, output)

def on_button_clicked(b):
    with output:
        print("\nDeine Bestellung wurde bestätigt und die Lebensmittel sind bereits auf dem Weg zu dir.\nVielen Dank und bis zum nächsten Mal!")

button.on_click(on_button_clicked)


```
